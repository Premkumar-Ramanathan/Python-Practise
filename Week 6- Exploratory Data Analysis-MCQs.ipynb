```python
column_name = 'LotFrontage'

# Calculate the upper bound based on a percentile (e.g., 95th percentile)
upper_percentile = 95
upper_bound = df[column_name].quantile(upper_percentile / 100)

# Replace values above the upper bound with the upper bound
df[column_name] = df[column_name].apply(lambda x: upper_bound if x > upper_bound else x)

# Print the modified DataFrame
print("DataFrame after capping upper-side outliers:")
print(df)
```

```python
ub=df["LotFrontage"].describe()["75%"]+1.5*(df["LotFrontage"].describe()["75%"]-df["LotFrontage"].describe()["25%"])
df["LotFrontage"]=np.where(df["LotFrontage"]>ub,ub,df["LotFrontage"])
```

```python
lower_percentile = 5
lower_bound = df['LotFrontage'].quantile(lower_percentile / 100)
df['LotFrontage'] = df['LotFrontage'].apply(lambda x: lower_bound if x < lower_bound else x)
print("DataFrame after flooring lower-side outliers:")
print(df)
```
