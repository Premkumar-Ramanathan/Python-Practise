<h2 style="color:Brown"> Arithmetic Operations

Data science involves large mathematical computations; hence, it is important to understand the basics of these computations. Basic computations are performed using basic operators as follows.

```python

print(6 + 7)  #Addition
print(6 - 7)  #Subtraction
print(6 * 7)  #Multiplication
print(6 / 7)  #Division
print(6 // 7) #Floor Division
print(6 % 7)  #Modulus
print(6 ** 7) #Exponential
```

What would you do if more than one operator is present in a single expression?
Here, the operator precedence will come into the picture.
Mathematical operators follow a precedence rule called PEDMAS where:  
P: Parenthesis
E: Exponents
M: Multiplication
D: Division
A: Addition
S: Subtraction

Reference - [Python operator precedence](https://docs.python.org/3.7/reference/expressions.html?#operator-precedence)

<h2 style="color:Brown"> How to find the data type for a variable in Python

```python
integer_num = 1
floating_num = 1.3
string = 'Mike'
boolean = True

print (type(integer_num))
print (type(floating_num))
print (type(string))
print (type(boolean))
```
