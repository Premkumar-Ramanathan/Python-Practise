```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
#!pip install matplotlib
#!pip install seaborn
```

### Bar Plot

```python
data = {'C':20, 'C++':15, 'Java':30,
        'Python':35}
courses = list(data.keys())
values = list(data.values())

plt.bar(courses,values)
```

## Figure Size

```python
data = {'C':20, 'C++':15, 'Java':30,
        'Python':35}
courses = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10, 5))
plt.bar(courses,values)
```

## Labelling

### xlabel & ylabel

```python
data = {'C':20, 'C++':15, 'Java':30,
        'Python':35}
courses = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10, 5))
plt.bar(courses,values)
plt.xlabel("Languages")
plt.ylabel("Users")
plt.show()
```

### title

```python
data = {'C':20, 'C++':15, 'Java':30,
        'Python':35}
courses = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10, 5))
plt.bar(courses,values)
plt.xlabel("Languages")
plt.ylabel("Users")
plt.title("Language Vs Users")
plt.show()
```

### xticks & yticks

```python
data = {'C':20, 'C++':15, 'Java':30,
        'Python':35}
courses = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10, 5))
plt.bar(courses,values)
plt.xticks(rotation =45)
plt.yticks(rotation =45)
plt.xlabel("Languages")
plt.ylabel("Users")
plt.show()
```

## Customization

# color

```python
data = {'C':20, 'C++':15, 'Java':30,
        'Python':35}
courses = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10, 5))
plt.bar(courses,values,color='gr')
plt.xticks(rotation =45)
plt.yticks(rotation =45)
plt.xlabel("Languages")
plt.ylabel("Users")
plt.show()
```

### align and edgecolor

```python
data = {'C':20, 'C++':15, 'Java':30,
        'Python':35}
courses = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10, 5))
plt.bar(courses,values,color='gr',align="center")
plt.xticks(rotation =45)
plt.yticks(rotation =45)
plt.xlabel("Languages")
plt.ylabel("Users")
plt.show()
```

```python
data = {'C':20, 'C++':15, 'Java':30,
        'Python':35}
courses = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10, 5))
plt.bar(courses,values,color='gr',align="center",edgecolor="b")
plt.xticks(rotation =45)
plt.yticks(rotation =45)
plt.xlabel("Languages")
plt.ylabel("Users")
plt.show()
```

### fontdict-fontsize

```python
data = {'C':20, 'C++':15, 'Java':30,
        'Python':35}
courses = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10, 5))
plt.bar(courses,values)
plt.xticks(rotation =45)
plt.yticks(rotation =45)
plt.xlabel("Languages",fontdict={"color":"g","size":12})
plt.ylabel("Users",fontdict={"color":"g","size":12})
plt.show()
```

## Seaborn

```python
sns.barplot(courses,values)
```

## set style

```python
sns.set_style({"grid.color": "b", "grid.linestyle": "--"})
sns.barplot(courses,values)
```

### ScatterPlot

```python
x =[5, 7, 8, 7, 2, 17, 2, 9,
    4, 11, 12, 9, 6]
 
y =[99, 86, 87, 88, 100, 86,
    103, 87, 94, 78, 77, 85, 86]
 
plt.scatter(x, y)
 
# To show the plot
plt.show()
```

### alpha, c and label

```python
x =[5, 7, 8, 7, 2, 17, 2, 9,
    4, 11, 12, 9, 6]
 
y =[99, 86, 87, 88, 100, 86,
    103, 87, 94, 78, 77, 85, 86]
 
plt.scatter(x, y, c ="blue",alpha=0.5,edgecolor="y")
plt.xlabel("x",fontdict={"color":"g","size":12})
plt.ylabel("y",fontdict={"color":"g","size":12})
 
# To show the plot
plt.show()
```

```python
sns.set_style("whitegrid",{"grid.color": "r", "grid.linestyle": "--"})
sns.scatterplot(x,y)
```
