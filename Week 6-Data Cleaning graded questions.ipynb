```python
data = {
    'Item': ['Item1', 'Item2', 'Item3'],
    '2023-06-15': [100, 200, 150],
    '2023-09-20': [80, 180, 120],
    '2023-12-10': [60, 160, 110],
    '2023-03-05': [90, 190, 140],
    '2023-07-30': [110, 210, 160]
}

df = pd.DataFrame(data)

# Convert column names (dates) to datetime objects
df.columns = pd.to_datetime(df.columns)

# Define a function to categorize dates into seasons
def categorize_season(date):
    month = date.month
    if 6 <= month <= 8:
        return 'Summer'
    elif 9 <= month <= 11:
        return 'Autumn'
    elif 12 <= month or month <= 2:
        return 'Winter'
    else:
        return 'Spring'

# Apply the categorize_season function to column names to obtain seasons
seasons = [categorize_season(date) for date in df.columns]

# Add a "Season" column to the DataFrame
df['Season'] = seasons

# Print the updated DataFrame
print(df)
```
