```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
plt.figure(figsize=(10,6))
sns.swarmplot(x='species',y='petal_length',data=iris)
plt.title('swarm plot')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='species',y='petal_length',data=iris)
plt.title('swarm plot')
plt.show()
```

```python
## stacked bar chart
A = [5,25,50,20]
B = [15,55,10,25]
xz = range(4)
plt.bar(xz,A,color='b')
plt.bar(xz,B,color='r',bottom=A)
plt.show()
```

```python
A = [5,25,50,20]
B = [15,55,10,25]
xz = range(4)
width = 0.4
plt.bar([x - width/2 for x in xz],A,width,color='b',label='A')
plt.bar([x + width/2 for x in xz],B,width,color='r',label='B')
plt.show()
```
