### Boxplot

#### It is used to see the quartile-wise distribution of a continuous variable
#### It is also used to see the outliers present in the data
#### It is used to see the quartile-wise distribution of a continuous variable against a categorical variable

```python
sns.boxplot(df["Fare"])
```

```python
df["Fare"].describe()
```

```python
for i in cont_cols:
    print(i)
    sns.boxplot(df[i])
    plt.show()
```

```python
## upper_whisker --> 75_per + 1.5*(75_per - 25_per)
## lower_whisker --> 25_per - 1.5*(75_per - 25_per)
## (75_per - 25_per)--> IQR (Inter-Quartile-Range)
```

## Bivariate Analysis

```python
sns.boxplot(df["Pclass"],df["Fare"])
```

```python
ls=[]
for i in cat_cols:
    for j in cont_cols:
        print(i,"Vs",j)
        ls.append((i,j))
        sns.boxplot(df[i],df[j])
        plt.show()
ls
```

## Barplot

#### It is used to see statistical distribution (mean/median/var/std/max/min etc)
#### of a continuous variable against a categorical variable

```python
sns.barplot(df["Pclass"],df["Fare"],ci=None)
## by default it shows mean
```

```python
ls=[]
for i in cat_cols:
    for j in cont_cols:
        print(i,"Vs",j)
        ls.append((i,j))
        sns.barplot(df[i],df[j],ci=None)
        plt.show()
ls
```
